```python
import numpy as np
import matplotlib.pyplot as plt

def fibonaci(n):
    if n == 1:
        return 1
    array = np.zeros(n)
    array[0], array[1] = 1,1
    for i in range(2,n):
        array[i] = array[i-1] + array[i-2]
    
    return array

fibonaci(2)
```

```python
# minimizacija funkcije jedne promenjlive 
# funkcija mora biit unimodalna nad intervalom[a, b]
# eps je trazena sirina intervala u kome se nalazi minimum

def fibonacijeva_metoda(a, b, eps):
    
    # 1. trazimo najmanji broj n koji zadovoljava uslov
    n = 1
    while ((b - a) / eps) > fibonaci(n):
        n = n + 1
        
    # 2. odredjujemo pocetne tacke
    x1 = a + fibonaci(n - 2) / fibonaci(n) * (b - a)
    x2 = a + b - x1
    
    # 3. iteracije
    for i in range(2, n+1):
        if func(x1)<=func(x2):
            b = x2
            x2 = x1
            x1 = a + b - x2
        else:
            a = x1
            x1 = x2
            x2 = a + b - x1
            
    if func(x1)<func(x2):
        xopt = x1
        fopt = func(x1)
    else:
        xopt = x2
        fopt = func(x2)
        
    return xopt,fopt,n
```
